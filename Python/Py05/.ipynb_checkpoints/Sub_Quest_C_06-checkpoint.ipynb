```python
## 은행계좌 만들기
class Account:
	
    account_quantity = 0    # 개설된 통장갯수(객체 갯수)(Q2)
    lst_account = []        # 개설된 통장 목록
    lst_history = []        # 입금/출금목록
 
	# 계좌 생성
    def __init__(self, *args):  
		# 변수선언
        self.bank_name   = 'SC'     # 은행명
        self.holder      = args[0]  # 예금주
        self.balance     = args[1]  # 잔고
        self.account_num = ''       # 계좌번호
        self.deposit_cnt = 0        # 입금횟수
		 
		# 11자리 계좌번호생성 (000-22-123356)
        self.account_num = self.create_account()
        
        # 개설 통장 수
        self.account_quantity += 1
        
        # 개설된 통장 정보 추가
        self.lst_account.append((self.bank_name, self.holder, self.account_num, self.balance))
			
		
	# 계좌번호생성
    def create_account(self):
        import random as r
		
		# 11자리 정수 랜덤생성
        num = str(r.randint(0, 99999999999)).zfill(11)
		
		# 999-99-999999 포멧
        acnt_num = f"{num[:3]}-{num[3:5]}-{num[5:]}"
        return acnt_num
    
    
    # 개설된 통장 수 출력 메소드(Q3)
    def get_account_num(self):
        print(f"지금까지 개설된 통장 수는 {self.account_quantity}개 입니다")    
    
    
    # 입금메소드(Q4)
    def deposit(self):
        while True:
            try:    
                balc =int(input('입금하실 금액을 입력하세요. >>> '))
                if balc > 0: 
                    # 입금금액 추가
                    self.balance += balc
                    print(f"{balc:,}원을 입금하여 총잔액 {self.balance:,}원이 되었습니다.")
                    
                    # 입금카운트+1
                    self.deposit_cnt += 1
                    
                    # 입금내역 history 저장
                    self.lst_history.append(("입금", balc, self.balance)) 
                    
                    # 이자 지급 대상여부 체크 및 지급(Q7)
                    if self.deposit_cnt == 5:
                        # 이자지급
                        self.balance += int(self.balance * 0.01)
                        self.deposit_cnt = 0  # 입금횟수 초기화
                        
                    break
                
                else:
                    print("1원이상 입력하세요.")    
            
            except Exception as e:
                print(f"잘못 입력하셨습니다.")
        
    
    # 출금 메소드(Q5)
    def withdraw(self):
        while True:
            try:       
                # 잔액이 0이면 종료
                if self.balance == 0:
                    print(f"{self.holder}님의 잔액이 0원으로 출금할 금액이 없습니다.")
                    return
                                
                # 출금 금액 입력/체크/처리    
                widr = int(input('출금하실 금액을 입력하세요. >>> '))
                if widr > 0: 
                    print(self.balance, widr)
                    # 잔액 < 출금금액 체크
                    if self.balance < widr:
                        print(f"잔액보다 출금금액이 큽니다. ({self.balance}<{self.widr})")
                
                    self.balance -= widr
                    print(f"{widr:,}원을 출금하여 총잔액 {self.balance:,}원이 되었습니다.")
                    
                    # 출금내역 history 저장
                    self.lst_history.append(("출금", widr, self.balance)) 
                
                    break;
                else:
                    print("1원이상 입력하세요.") 
            except Exception as e:
                print(f"잘못 입력하셨습니다.")                       
    
    
    # 정보출력 메서드 (Q6/Q9)
    # args 입력값 부재시 전체 출력
    # args 입력값이 있으면 잔고가 해당금액 이상인 예금주 정보 출력 
    def display_info(self, *args):
  
        # print("-"*70)
        for i in self.lst_account:
            if args: 
                if i[3] >= args[0]:
                    print(f"  은행이름:{i.bank_name}, 예금주:{i.holder}, 계좌번호:{i.account_num}, 잔고금액:{i.balance:,}원")
            else:
                print(f"  은행이름:{i.bank_name}, 예금주:{i.holder}, 계좌번호:{i.account_num}, 잔고금액:{i.balance:,}원")
                
        # print("-"*70)
       
    
    # 통장 거래내역 출력
    def print_account(self):
        print(f"계좌번호 : {self.account_num}")
        while True:    
            p = int(input("출력할 거래내역을 선택하세요. 0.전체 1.입금 2.출금 >>> "))
            # 전체
            if p == 0:
                for hist in self.lst_history:
                    print(f" {hist[0]} : {hist[1]:,}원     잔액 : {hist[2]:,}원")
                break
            # 입금
            elif p == 1:
                self.deposit_histoty()
                break
            # 출금
            else:    
                self.withdraw_history()
                break
                
                
    # 입금 내역 출력(Q10)    
    def deposit_histoty(self):
        for hist in self.lst_history:
            if hist[0] == "입금":
                print(f"입금 : {hist[1]:,}원     잔액 : {hist[2]:,}원")
        
    # 출금 내역 출력(Q10)    
    def withdraw_history(self):
        for hist in self.lst_history:
            if hist[0] == "출금":
                print(f"출금 : {hist[1]:,}원     잔액 : {hist[2]:,}원")      
        
```

```python
a = Account('아이펠', 10000)
b = Account('모두연', 10000000)
c = Account('파이썬', 1000)
```

```python
a.display_info()
```
